```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

PLOT Exercicio 1.2 a)

```python
date = pd.to_datetime(["02/10/2023", "03/10/2023", "04/10/2023", "05/10/2023", "06/10/2023"], format="%d/%m/%Y")
abertura = [132.15, 134.93, 133.66, 136.13, 134.94]

df = pd.DataFrame({'Data': date, 'Abertura': abertura})
df['DataFormatada'] = df['Data'].dt.strftime('%d/%m/%Y')

fig, ax = plt.subplots(figsize=(10, 5))

ax.set_title("Gráfico de Dispersão dos Valores de Abertura")
ax.plot(abertura, label="Valor abertura")
ax.set_xlabel("Data")
ax.set_ylabel("Valor de Abertura")
ax.set_xticks(range(len(date)))
ax.set_xticklabels(df['DataFormatada'])
ax.set_yticks(range(130, 141))
ax.legend()
ax.grid(True)
plt.show()
```

```python
dias = np.array((date - date.min()).days + 1)
print(dias)

alphal2 = 132.328
betal2 = 0.678

valores_ajustadosl2 = alphal2 + betal2 * dias

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(abertura, label="Valor abertura")
ax.plot(valores_ajustadosl2, color='red', label="Reta de Regressão (L2)", linestyle='--')
ax.set_title('Gráfico de Dispersão com Regressão Linear')
ax.set_xlabel('Data')
ax.set_ylabel('Valor de Abertura')
ax.set_xticks(range(len(date)))
ax.set_xticklabels(df['DataFormatada'])
ax.set_yticks(range(130, 141))
ax.grid(True)
ax.legend()
plt.show()
```

```python
alphal1 = 131.740
betal1 = 0.640

valores_ajustadosl1 = alphal1 + betal1 * dias

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(abertura, label="Valor abertura")
ax.plot(valores_ajustadosl1, color='red', label="Reta de Regressão (L1)", linestyle='--')
ax.set_title('Gráfico de Dispersão com Regressão L1')
ax.set_xlabel('Data')
ax.set_ylabel('Valor de Abertura')
ax.set_xticks(range(len(date)))
ax.set_xticklabels(df['DataFormatada'])
ax.set_yticks(range(130, 141))
ax.legend()
ax.grid()
plt.show()

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(abertura, label="Valor abertura")
ax.plot(valores_ajustadosl1, color='orange', label="Reta de Regressão (L1)", linestyle='--')
ax.plot(valores_ajustadosl2, color='red', label="Reta de Regressão (L2)", linestyle='--')
ax.set_title('Gráfico de Dispersão com Regressão L1 e L2')
ax.set_xlabel('Data')
ax.set_ylabel('Valor de Abertura')
ax.set_xticks(range(len(date)))
ax.set_xticklabels(df['DataFormatada'])
ax.set_yticks(range(130, 141))
ax.legend()
ax.grid()
plt.show()
```
